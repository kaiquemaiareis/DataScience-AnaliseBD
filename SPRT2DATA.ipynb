```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
# Exibe os nomes das colunas
print(df.columns)
```

```python
# Obtem o numero de entradas, tipos de dados e valores nulos
print(df.info())
# Numero de valores nulos em cada coluna
df.isnull().sum()
```

```python
# Estatisticas descritivas gerais
df.describe()
```

```python
# Contagem por estado (uf)
contagem_uf = df['uf'].value_counts()

# Contagem por sexo
contagem_sexo = df['sexo'].value_counts()

# Contagem por faixa etaria
contagem_faixa_etaria = df['faixa_etaria'].value_counts()

# Visualizar os resultados
contagem_uf, contagem_sexo, contagem_faixa_etaria
```

```python
# Grafico de barras para distribuicao por estado
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='uf')
plt.title('Distribuição de Óbitos por Estado')
plt.xlabel('Estado')
plt.ylabel('Contagem')
plt.xticks(rotation=90)
plt.show()

# Grafico de barras para distribuicao por sexo
plt.figure(figsize=(8, 3))
sns.countplot(data=df, x='sexo')
plt.title('Distribuição de Óbitos por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()

# Grafico de barras para distribuicao por faixa etaria
plt.figure(figsize=(9, 4))
sns.countplot(data=df, x='faixa_etaria')
plt.title('Distribuição de Óbitos por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()

# Grafico de barras para distribuicao por tipo de doenca
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='tipo_doenca')
plt.title('Distribuição de Óbitos por Tipo de Doença')
plt.xlabel('Tipo de Doença')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grafico de barras agrupado por tipo de doenca e faixa etaria
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='faixa_etaria', hue='tipo_doenca', palette='Set2')
plt.title('Número de Óbitos por Tipo de Doença e Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.legend(title='Tipo de Doença')
plt.tight_layout()
plt.show()
```
